# **Building the Words dictionary**

```python
words_dictionary = pd.read_csv('/content/drive/MyDrive/words_dictionary.csv', encoding='latin-1')

words_dictionary.sort_values('Total', ascending = False)
```

```python
cond1 = words_dictionary['Total'] > 5
sum(cond1)
```

```python
words_dictionary = words_dictionary[cond1]
```

```python
#words_dictionary = words_dictionary[~ (words_dictionary['Unnamed: 0'] == 'covid')]
```

```python
vocabulary = set(words_dictionary.iloc[:,0])
data = np.array(words_dictionary.iloc[:,1:6])
suma = sum(data)

for i in range(0,5):
  for j in range(0,len(data)):
    data[j,i] = data[j,i]/suma[i]
    if data[j,i] == 0:
      data[j,i] = 1/(suma[i] + len(vocabulary))
```

```python
for i in range(0,5):
  words_dictionary[i+1] = data[:,i]
```

```python
words_dictionary.sort_values('Total', ascending = False)
```
