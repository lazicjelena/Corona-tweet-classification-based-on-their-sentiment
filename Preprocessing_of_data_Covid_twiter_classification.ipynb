```python
def unique_column_values(df,s):

  #input arguments:
  # df - data frame
  # s -string, column name

  # check how many unique values there are in the column s of the dataframe df
  # return how many times every unique values appears in columns of the dataframe df

  unique_values = df[s].unique()
  values = []
  for i in unique_values:
    print('{0}  {1}'.format(i , len(df[df[s]==i])))
    values.append(len(df[df[s]==i]))

  return values
```

Removing stopwords with NLTK.corpus library. Stopwords are:
> I, me, my, myself, we, our, you've, you'll, you'd, your, she's, her, hers, herself....

```python
# download the stopwords from NLTK
nltk.download('stopwords')
```

```python
#Import the english stop words list from NLTK
stopwords_english = stopwords.words('english') 
stemmer = PorterStemmer() # instantiate stemmer
```

```python
df_train["TweetTokens"] = "" # add additional column to dataframe
df_test["TweetTokens"] = ""

# instantiate tokenizer class
tokenizer = TweetTokenizer(preserve_case=False, strip_handles=True,reduce_len=True)
```

```python
def is_ascii(s):
    # removes all words consisting with non-alphabet char or numbers
    return all(ord(c) < 123 and ord(c) > 96 for c in s) 
```

Preprocessing of each tweet:
1.   remove old style retweet text "RT"
2.   replace hiperlink with postojaojelink
3.  remove hashtags, only removing the hash # sign from the word

Preprocess every single tweet.



```python
for i in range(0,len(df_train)):

  tweet = df_train.at[i,'OriginalTweet']
  tweet =  re.sub(r'^RT[\s]+', retweet_string, tweet)
  tweet = re.sub(r'https?://[^\s\n\r]+', hiperlink_string, tweet)
  tweet = re.sub(r'#', '', tweet)
  tweet = re.sub(r"@\w+", mention_string,tweet) 
  # tokenize tweets
  df_train.at[i,'TweetTokens'] = tokenizer.tokenize(tweet)

  tweets_clean = []
  for word in df_train.at[i,'TweetTokens']: # Go through every word in your tokens list
    if word == 'coronavirus':
      word = 'covid'  # coronavirus and covid are semantically the same
    if (word not in stopwords_english and word not in string.punctuation and is_ascii(word)):
      tweets_clean.append(stemmer.stem(word)) #stemming word
  if len(tweets_clean) > 0:
    df_train.at[i,'TweetTokens'] = tweets_clean
  else:
    df_train.at[i,'TweetTokens'] = ''
```

```python
for i in range(0,len(df_test)):

  tweet = df_test.at[i,'OriginalTweet']
  tweet =  re.sub(r'^RT[\s]+', retweet_string, tweet)
  tweet = re.sub(r'https?://[^\s\n\r]+', hiperlink_string, tweet)
  tweet = re.sub(r'#', hashtag_string, tweet)
  tweet = re.sub(r"@\w+", mention_string,tweet) 
  # tokenize tweets
  df_test.at[i,'TweetTokens'] = tokenizer.tokenize(tweet)

  tweets_clean = []
  for word in df_test.at[i,'TweetTokens']: # Go through every word in your tokens list
    if word == 'coronavirus':
      word = 'covid'  # coronavirus and covid are semantically the same
    if (word not in stopwords_english and word not in string.punctuation and is_ascii(word)):
      tweets_clean.append(stemmer.stem(word)) #stemming word
  if len(tweets_clean) > 0:
    df_test.at[i,'TweetTokens'] = tweets_clean
  else:
    df_test.at[i,'TweetTokens'] = ''
```

Remove all tweets left with no tokens.

# **Building the Words dictionary**

```python
# calculate word frequencies

words_dictionary = pd.DataFrame({'Extremely Positive':[],'Positive':[],'Neutral':[],'Negative':[],'Extremely Negative':[],'Total':[]})
words_set = set()

for i in range(0,len(df_train)):
  for word in df_train.iloc[i]['TweetTokens']:
    if not(word in words_set):
      words_set.add(word)
      words_dictionary.loc[word] = [0,0,0,0,0,0]
    words_dictionary.loc[word][df_train.iloc[i]['Sentiment']] += 1
    words_dictionary.loc[word]['Total'] += 1
```

```python
words_dictionary.sort_values('Total', ascending = False)
```

```python
cond1 = words_dictionary['Total'] > 5
print(f"There are {sum(cond1)} words left in dictionary after removing rare words.")
words_dictionary = words_dictionary[cond1]
```

```python
words_dictionary.to_csv('/content/drive/MyDrive/words_dictionary.csv', index = True)
```

# Calculate probabilities for Naive Bayes classification

```python
vocabulary = set(words_dictionary.index)
data = np.array(words_dictionary.iloc[:,0:5])
suma = sum(data)

for i in range(0,5):
  for j in range(0,len(data)):
    data[j,i] = data[j,i]/suma[i]
    if data[j,i] == 0:
      data[j,i] = 1/(suma[i] + len(vocabulary))
```

```python
for i in range(0,5):
  words_dictionary[i+1] = data[:,i]
```
